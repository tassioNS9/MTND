Uma unidade de controle que contém um registrador de status e uma tabela de transição predeterminada que informa à máquina o estado atual, se deve ler ou escrever símbolos na fita e o estado de destino

### Implementação

Foi determinado um simulador de uma máquina de Turing não determínistica onde as entradas são dada pelos estados, os alfabetos de entrada e fita, o simbolo de divisão e o simbolo branco, além disso as entradas da quantidade de transições  e as transições a serem lidas, juntamente com os estados iniciais e finais e as palavras a serem computadas. As transformações são especificadas como quíntuplas <a, b, c, d, e> onde "a" é o estado de origem, "b" é o caractere a ser lido e "c" é o estado de origem. destino, "d" é o símbolo a ser escrito e, finalmente, "e" é a direção, parado (P), esquerda (E) e direita (D). O programa recebe os dados descritos no parágrafo anterior como uma string, com exceções Número de conversões para números inteiros. A principal informação a mencionar é a transformação armazenada usando a estrutura de dados do dicionário, onde a chave consiste no estado de origem e no caractere lido, e o valor é o estado de destino, Símbolos e direções. O programa foi realizado utilizando a linguagem python por ser mais fácil desenvolvimento e possuir simplicidade e robustez. A ferramenta de desenvolvimento utilizada foi o visual estudio.

```python

estds = input().split() 
alfabeto_Entrada=input().split() 
alfabeto_Fita = input().split()


divisor = input()  
valor_branco = input() 
qtdTransicoes=int(input())
transicoes=dict()
contador = 0

while(contador < qtdTransicoes):
    quintupla = input().split() 
    key = (quintupla[0], quintupla[1])
    if key not in transicoes:
        transicoes[key]=([[quintupla[2], quintupla[3], quintupla[4]]])
    else:
        transicoes[key].append([quintupla[2], quintupla[3], quintupla[4]])
    contador = contador + 1


estdInicial = input()
estdsFinais = input().split()
palavras = input().split()
```

A parte principal do algoritmo é a verificação da aceitação ou não da palavra pela máquina descrita. A palavra é repartida em símbolos e de acordo o estado inicial verifica se o conjunto estado símbolo corresponde a uma das chave contidas no alfabeto de transições. A fita é determinada por uma lista composta do simbolo divisor, a palavra recebida pela entrada e o simbolo branco para cada uma dessas entradas, além disso foi determinada uma pilha auxiliar ,portanto, se a chave calculada atual estiver no dicionário, atribua a palavra dada a este Empilha a configuração atual com as posições do cabeçote e da fita. Então temos não determinismo onde cada computação executa uma pesquisa de estado e símbolo atual. Repita este processo até que a palavra chegue ao fim e a máquina esteja no estado.

```python
valorAux = 1


for palavra in palavras:
    fita=list((divisor+palavra+valor_branco))
    pilha=[(estdInicial, valorAux, fita)]
    aceitacao=False
    while len(pilha)>0:
        p=pilha.pop()
        estdAtual=p[0]
        index_p=p[1]
        fita_p=p[2][:]
        key_p=(estdAtual,fita_p[index_p])
        if key_p not in transicoes and estdAtual in estdsFinais:
            aceitacao=True
            break
        if key_p in transicoes:          
            for a in transicoes[(key_p)]:               
                estdAtual = a[0] 
                fita_p[index_p] = a[1]
                direcionamento = a[2]    
                if direcionamento == 'E':                   
                    pilha.append((estdAtual, index_p-1, fita_p))
                elif direcionamento == 'D':
                    if (index_p+1)==len(fita):
                        fita_p.append(valor_branco)
                    pilha.append((estdAtual, index_p+1, fita_p))
                elif direcionamento == 'P':
                    pilha.append((estdAtual, index_p, fita_p))
                    
    if aceitacao:
        print("S")
    else:
        print("N")
```
